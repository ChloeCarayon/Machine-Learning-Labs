```python
# FINAL IDEE

def meanAbsoluteError(X,w,Y):
    den = 1/len(Y)
    Sum_er = sum(abs(X.dot(w) - Y)) 
    return Sum_er*den

def rootMeanSquaredError(X,w,Y): 
    den = 1/len(Y)
    # revoir absolut = abs(X.dot(w) - Y).T
    Sum_er = sum(abs(X.dot(w) - Y)**2)
    total =  (Sum_er*den)**(1/2)
    return total

def r2(X,w,Y): 
    SSres = sum(abs(X.dot(w) - Y)**2)
    SStot = sum(abs(Y - np.mean(Y))**2)
    total = 1 - SSres /SStot
    return total

    
```
